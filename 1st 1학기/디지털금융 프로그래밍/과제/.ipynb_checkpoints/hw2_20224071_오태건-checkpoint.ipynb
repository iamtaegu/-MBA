```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('historical simulation.xlsx', header=8)
df.head()
```

```python
sp500 = 4000
ftse100 = 3000
cac40 = 1000
nikkei225 = 2000
tot = sp500 + ftse100 + cac40 + nikkei225

df['FTSE100($)']  = df.loc[:, 'FTSE 100']*df.loc[:, 'USD/GBP']
df['CAC400($)'] = df.loc[:, 'CAC40'] / df.loc[:, 'EUR/USD']
df['Nikkei225($)'] = df.loc[:, 'Nikkei 225'] / df.loc[:, 'JPY/USD']
df = df.drop(columns=['Date','FTSE 100', 'CAC40', 'Nikkei 225', 'USD/GBP', 'EUR/USD', 'JPY/USD'], axis=1)
df = df.rename(columns={'Day':'Scenario'})
```

```python
for i in range(len(df)-1):
    df.loc[i, 'S&P500'] = sp500 * df.loc[i+1, 'S&P500'] / df.loc[i, 'S&P500']
    df.loc[i, 'FTSE100($)'] = ftse100 * df.loc[i+1, 'FTSE100($)'] / df.loc[i, 'FTSE100($)'] 
    df.loc[i, 'CAC400($)'] = cac40 * df.loc[i+1, 'CAC400($)'] / df.loc[i, 'CAC400($)'] 
    df.loc[i, 'Nikkei225($)'] = nikkei225 * df.loc[i+1, 'Nikkei225($)'] / df.loc[i, 'Nikkei225($)'] 
    portfolioValue = 0
    for j in range(1, 5):
        portfolioValue += df.iloc[i, j]
    df.loc[i, 'portfolio value'] = portfolioValue
    df.loc[i, 'P/L'] = df.loc[i, 'portfolio value'] - tot
    df.loc[i, 'P/L(%)'] = df.loc[i, 'P/L'] / 100 
    
df = df.drop(index=500)
df.loc[:, 'Scenario'] = df.loc[:, 'Scenario'] + 1
df = df.round(2)
df = df.sort_values('P/L')
df.set_index('Scenario')
```
