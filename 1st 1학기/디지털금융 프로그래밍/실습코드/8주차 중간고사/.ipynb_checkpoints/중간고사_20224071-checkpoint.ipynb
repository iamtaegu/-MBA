# 문제3

```python
wealth = 1000000
riskless = 0.06
annual = 0.08
sd = 0.2
proportion = 0.7
drawdown = 150000
```

```python
np.random.seed(0)
year=np.arange(1,11)
df = pd.DataFrame(index=year, columns=list('BCDEFGHI'))
df.loc[:,'H'] = drawdown
df.loc[:,'E'] = np.random.randn(10)

df.loc[1, 'B'] = wealth
df.loc[1, 'C'] = wealth * proportion
df.loc[1, 'D'] = wealth * (1-proportion)
df.loc[1, 'F'] = np.exp( (annual-0.5*np.power(2, sd)) + sd*df.loc[1,'E'])
df.loc[1, 'G'] = df.loc[1, 'C'] * df.loc[1, 'F'] + df.loc[1, 'D'] * np.exp(riskless)
df
```

```python
for i in range(2, len(df)+1):
    df.loc[i, 'B'] = df.loc[i-1, 'G'] - df.loc[i-1, 'H']
    df.loc[i, 'C'] = df.loc[i, 'B'] * proportion 
    df.loc[i, 'D'] = df.loc[i, 'B'] * (1-proportion)
    df.loc[i, 'F'] = np.exp( (annual-0.5*np.power(2, sd)) + sd*df.loc[i,'E'])
    df.loc[i, 'G'] = df.loc[i, 'C'] * df.loc[i, 'F'] + df.loc[i, 'D'] * np.exp(riskless)
df.loc[10, 'I'] = df.loc[10, 'G'] - df.loc[10, 'H'] 
```

# 문제5 

```python
def PositiveWealth(wealth, riskless, annual, sd, proportion, drawdown, run):
    positiveCnt = 0
    np.random.seed(0)
    for cnt in range(run):
        year=np.arange(1,11)
        df = pd.DataFrame(index=year, columns=list('BCDEFGHI'))
        df.loc[:,'E'] = np.random.randn(10)
        df.loc[:,'H'] = drawdown

        df.loc[1, 'B'] = wealth
        df.loc[1, 'C'] = wealth * proportion
        df.loc[1, 'D'] = wealth * (1-proportion)
        df.loc[1, 'F'] = np.exp( (annual-0.5*np.power(2, sd)) + sd*df.loc[1,'E'])
        df.loc[1, 'G'] = df.loc[1, 'C'] * df.loc[1, 'F'] + df.loc[1, 'D'] * np.exp(riskless)

        for i in range(2, len(df)+1):
            df.loc[i, 'B'] = df.loc[i-1, 'G'] - df.loc[i-1, 'H']
            df.loc[i, 'C'] = df.loc[i, 'B'] * proportion 
            df.loc[i, 'D'] = df.loc[i, 'B'] * (1-proportion)
            df.loc[i, 'F'] = np.exp( (annual-0.5*np.power(2, sd)) + sd*df.loc[i,'E'])
            df.loc[i, 'G'] = df.loc[i, 'C'] * df.loc[i, 'F'] + df.loc[i, 'D'] * np.exp(riskless)
        df.loc[10, 'I'] = df.loc[10, 'G'] - df.loc[10, 'H'] 
        if (df.loc[10, 'I'] > 0):
            positiveCnt += 1

    print(f"Successful runs = {round(positiveCnt/1000 * 100, 2)}%")
```

```python
wealth = 1000000
riskless = 0.08
annual = 0.10
sd = 0.4
proportion = 0.4
drawdown = 100000
run = 1000
PositiveWealth(wealth, riskless, annual, sd, proportion, drawdown, run)
```

# 문제6

```python
import math
```

```python
def WealthWithCushion(wealth, riskless, annual, sd, proportion, drawdown, run, cushion):
    outList = []
    np.random.seed(0)
    for cnt in range(run):
        year=np.arange(1,11)
        df = pd.DataFrame(index=year, columns=list('BCDEFGHI'))
        df.loc[:,'E'] = np.random.randn(10)
        df.loc[1,'H'] = drawdown

        df.loc[1, 'B'] = wealth
        df.loc[1, 'C'] = wealth * proportion
        df.loc[1, 'D'] = wealth * (1-proportion)
        df.loc[1, 'F'] = np.exp( (annual-0.5*np.power(2, sd)) + sd*df.loc[1,'E'])
        df.loc[1, 'G'] = df.loc[1, 'C'] * df.loc[1, 'F'] + df.loc[1, 'D'] * np.exp(riskless)
        for i in range(2, len(df)+1):
            df.loc[i, 'B'] = df.loc[i-1, 'G'] - df.loc[i-1, 'H']
            df.loc[i, 'C'] = df.loc[i, 'B'] * proportion 
            df.loc[i, 'D'] = df.loc[i, 'B'] * (1-proportion)
            df.loc[i, 'F'] = np.exp( (annual-0.5*np.power(2, sd)) + sd*df.loc[i,'E'])
            df.loc[i, 'G'] = df.loc[i, 'C'] * df.loc[i, 'F'] + df.loc[i, 'D'] * np.exp(riskless)
            if (df.loc[i, 'G'] > cushion*drawdown): 
                df.loc[i, 'H'] = drawdown
            else: 
                df.loc[i, 'H'] = df.loc[i, 'G']/3
        df.loc[10, 'I'] = df.loc[10, 'G'] - df.loc[10, 'H'] 
        outList.append(df.loc[10, 'I'])
    print(f"Result mean= {np.mean(outList)} sd= {math.sqrt(np.var(outList))}")
       
            
```

```python
wealth = 1000000
riskless = 0.08
annual = 0.10
sd = 0.4
proportion = 0.4
drawdown = 100000
run = 1000
cushion = 3
WealthWithCushion(wealth, riskless, annual, sd, proportion, drawdown, run, cushion)
```
