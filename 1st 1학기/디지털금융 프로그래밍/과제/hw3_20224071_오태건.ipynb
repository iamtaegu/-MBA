```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.tseries.offsets import *
from scipy import stats

crsp = pd.read_csv('crsp_d(2005).csv')
crsp['date'] = pd.to_datetime(crsp['date'])
crsp.dropna(subset=['ret'], inplace=True)
crsp.sort_values(by=['permno', 'date'])
crsp['year'] = crsp['date'].dt.year
crsp['month'] = crsp['date'].dt.month
crsp.head()
```

```python
crsp = crsp.groupby(by=["permno", "year", "month"], as_index=False).max()
crsp = crsp.drop(columns='date')
crsp = crsp.rename(columns={'ret':'maxret'})
crsp
```
