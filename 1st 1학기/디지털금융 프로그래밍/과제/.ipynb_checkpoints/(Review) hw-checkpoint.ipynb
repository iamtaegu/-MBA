# HW1

```python
import numpy as np
import pandas as pd

V = 10000
F = 0.85
M = 4
rate = 0.02 / 12

df = pd.read_excel('CPPI.xlsx')
df['ret'] = df['sp500'].pct_change()
df['stock100'] = df['ret'] + 1
df.loc[0, 'stock100'] = V
df['stock100']  = df['stock100'].cumprod()
df.head(1)
```

```python
# 초기화
df.loc[0, 'CPPI'] = V
df.loc[:, 'floor'] = V * F
df.loc[0, 'cushion'] = df.loc[0, 'CPPI'] - V * F
df.loc[0, 'stock'] = df.loc[0, 'cushion'] * M
df.loc[0, 'CMA'] = df.loc[0, 'CPPI'] - df.loc[0, 'stock']
df.loc[0, 'ratio'] = df.loc[0, 'stock'] / df.loc[0, 'CPPI']
```

```python
df.head(1)
```

```python
for i in df.index:
    if i == 0:
        continue
    
    #CPPI
    df.loc[i, 'CPPI'] = df.loc[i-1, 'stock']*(1+df.loc[i, 'ret']) + df.loc[i-1, 'CMA'] * (1 + rate)
    #cushion 
    df.loc[i, 'cushion'] = max(df.loc[i, 'CPPI'] - V * F, 0)
    #stock
    df.loc[i, 'stock'] = min(df.loc[i, 'cushion']*M, df.loc[i, 'CPPI'])
    #CMA
    df.loc[i, 'CMA'] = df.loc[i, 'CPPI'] - df.loc[i, 'stock']
    #ratio
    df.loc[i, 'ratio'] = df.loc[i, 'stock'] / df.loc[i, 'CPPI'] 
```
